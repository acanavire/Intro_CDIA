#### Series

- Es un objeto similar a un **vector unidimensional**.

#### DataFrame

- Representa una estructura de datos tabular, similar a una hoja de c谩lculo, que contiene una colecci贸n ordenada de columnas, cada una de las cuales puede ser de un tipo de valor diferente (num茅rico, cadena, booleano, etc.).

Vamos a construir dos instancias de Series:

- `superficie_s`, a partir de un diccionario. Retomamos el ejemplo anterior.

- `poblaci贸n_s`, a partir de dos listas

Luego construimos el DataFrame `provincia_df` a partir de las dos Series.

````python
# Diccionario en python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km虏] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'C贸rdoba': 164707.8, 'Salta': 155340.5,'Tucum谩n': 22592.1,'La Rioja': 91493.7}

# Serie desde dict y le agregamos el nombre
superficie_s = pd.Series(superficie_dict, name = 'superficie')

# Serie
superficie_s
````

````python
# Diccionario en python formado por datos de provincias de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de poblaci贸n [km虏] asociado.

# Crear lista 1
provincia_lista = ['Buenos Aires','Salta','C贸rdoba', 'Tucum谩n', 'La Rioja']

# Crear lista 2
provincia_pop = [17541141, 1424397, 3760450, 1694656, 393531]

# Serie de provincias desde dos listas
poblacion_s = pd.Series(provincia_pop, index= provincia_lista, name= 'poblacion')

# Mostrar Serie con el nombre poblacion
poblacion_s
````

````python
# Mirar atributos de Series
print('Colecci贸n de valores:',poblacion_s.values)
print('Colecci贸n de etiquetas (indice):',poblacion_s.index)
````

Para crear un DataFrame desde un diccionario, tener en cuenta que las claves son las etiquetas de las columnas y los valores son los valores de las colecciones asociadas. Para el ejemplo son los valores de las series creadas previamente.

````python
# Crear Dataframe
provincia_df = pd.DataFrame( {'Poblaci贸n':poblacion_s, 'Superficie(km2)':superficie_s}  )
````

### Tipos de Variables

En estad铆stica las variables, las que se pueden variar, se clasifican en diferentes tipos seg煤n la naturaleza de los datos que representan. Aqu铆 un resumen con ejemplos:

**Variables Cualitativas (o Categ贸ricas)**

Son aquellas que representan caracter铆sticas o categor铆as que no pueden ser medidas con n煤meros. Se dividen en dos tipos:

* Nominales: No tienen un orden espec铆fico.
  * Ejemplo: Color de ojos (azul, verde, marr贸n), estado civil (soltero, casado, divorciado).
* Ordinales: Tienen un orden o jerarqu铆a, pero las diferencias entre valores no son necesariamente iguales.
  * Ejemplo: Nivel de satisfacci贸n (bajo, medio, alto), nivel educativo (primaria, secundaria, universidad).

**Variables Cuantitativas (o Num茅ricas)**

Son aquellas que representan cantidades, medidas con n煤mero y permiten realizar operaciones matem谩ticas. Se dividen en dos tipos:

* Discretas: Toman un valor finito de valores. Son valores enteros contables (no pueden tomar valores intermedios).
  * Ejemplo: N煤mero de integrantes en el hogar (0, 1, 2, 3), cantidad de autos en un estacionamiento.
* Continuas: Pueden tomar cualquier valor dentro de un rango n煤merico, incluyendo decimales.
  * Ejemplo: Altura de una persona (1.75 m), temperatura en grados Celsius (23.5掳C).


En visualizaci贸n, el tipo de variable determina la gr谩fica adecuada para su representaci贸n.

Existen diversas formas de representar datos de manera visual. Algunas de las m谩s utilizadas son:

* Gr谩ficos de barras: Ideales para comparar valores de variables categ贸ricas.

* Histogramas: tiles para mostrar la distribuci贸n de una variable num茅rica, es decir c贸mo se distribuyen los diferentes valores de esa variable a lo largo de las observaciones.

* Diagramas de dispersi贸n (scatter plots): Se emplean para analizar la relaci贸n entre dos variables num茅ricas.

* Gr谩ficos de l铆neas: Representan tendencias a lo largo del tiempo.

* Boxplots: Muestran la distribuci贸n y posibles valores at铆picos en un conjunto de datos.
