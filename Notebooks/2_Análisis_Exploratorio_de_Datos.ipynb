**Pandas**

- Provee **operaciones de agrupación por columnas, filtros y sumarizaciones**.

- Estas características hacen de Pandas una librería **indispensable para las tareas de limpieza y preprocesamiento de datos**.

#### Series
---
- Es un objeto similar a un **vector unidimensional**.
- Colección de valores que mantiene un orden. Los valores pueden ser de cualquier tipo de datos, como números enteros, cadenas, objetos de Python, etc.
- Esta formado por una colección de valores y una colección de etiquetas de estos valores denominado **índice** (index).



image.png


La serie más simple se forma a partir de una lista de datos.

Si necesitas reforzar sobre las estructuras de datos en Python, mirar la notebook **Fundamentos** de este curso o click en [Estructura de datos de la documentación oficial de Python](https://docs.python.org/es/3/tutorial/datastructures.html).

````python
# Pasar como argumento la lista = [4, 17, -51, 3]
s = pd.Series([4, 17, -51, 3])
s
````

Como no especificamos un índice para los datos, se crea uno predeterminado que consta de los números enteros del 0 al N - 1 (donde N es la longitud de los datos). Para el ejemplo la serie tiene una colección de N=4 valores y el indice es una colección de enteros de 0 a 3.

Para acceder a la colección de valores y al indice considerar los atributos de la serie `value` e `index` respectivamente:

````python
# Colección de valores
s.values
````

````python
# Colección de etiquetas. Indice
s.index
````

Si tiene datos contenidos en un diccionario de Python, puede crear una serie a partir de ellos pasando el diccionario:

````python
# Diccionario en Python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km²] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'Córdoba': 164707.8, 'Salta': 155340.5,'Tucumán': 22592.1,'La Rioja': 91493.7}

# Serie desde dict
superficie_s = pd.Series(superficie_dict)

# Serie
superficie_s
````

Cuando solo se pasa un diccionario, el índice de la serie resultante tendrá las claves del diccionario en el orden generado.


````python
# Crear serie desde un diccionario e indice

# Crear lista de provincias
provincias = ['Buenos Aires', 'Córdoba', 'Salta', 'Misiones']

# Crear serie
superficie_s2 = pd.Series(superficie_s, index=provincias)

# Mostrar
superficie_s2
````

En este caso, se colocaron 3 valores encontrados en superficie_s en las ubicaciones correspondientes, pero como no se encontró ningún valor para 'Misiones', aparece como `NaN` (Not a Number), lo que en Pandas se considera para marcar valores faltantes o NA.

Usaremos los términos "faltante" o "NA" para referirnos a los datos faltantes. Las funciones `isnull` y `notnull` en Pandas se usan para detectar datos faltantes:

````python
# Usar la función isnull de Pandas. Se pasa como argumento una serie.
# Se evalúa si el dato en DataFrame es dato faltante.
pd.isnull(superficie_s2)
````

````python
# Usar la función notnull de Pandas. Se pasa como argumento una serie.
# Se evalúa si el dato en DataFrame no es dato faltante.
pd.notnull(superficie_s2)
````

La serie también tiene estos métodos de instancia.

````python
# Función isnull de la serie
superficie_s2.isnull()
````

#### DataFrame
---
- Representa una estructura de datos tabular, similar a una hoja de cálculo, que contiene una colección ordenada de columnas, cada una de las cuales puede ser de un tipo de valor diferente (numérico, cadena, booleano, etc.).
- El `DataFrame` tiene índices de columnas y filas.
- Cada columna de un `DataFrame` es una `Serie`.



image.png

Vamos a construir dos instancias de Series:

- `superficie_s`, a partir de un diccionario. Retomamos el ejemplo anterior.

- `población_s`, a partir de dos listas

Luego construimos el DataFrame `provincia_df` a partir de las dos Series.

````python
# Diccionario en python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km²] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'Córdoba': 164707.8, 'Salta': 155340.5,'Tucumán': 22592.1,'La Rioja': 91493.7}

# Serie desde dict y le agregamos el nombre
superficie_s = pd.Series(superficie_dict, name = 'superficie')

# Serie
superficie_s
````

````python
# Diccionario en python formado por datos de provincias de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de población [km²] asociado.

# Crear lista 1
provincia_lista = ['Buenos Aires','Salta','Córdoba', 'Tucumán', 'La Rioja']

# Crear lista 2
provincia_pop = [17541141, 1424397, 3760450, 1694656, 393531]

# Serie de provincias desde dos listas
poblacion_s = pd.Series(provincia_pop, index= provincia_lista, name= 'poblacion')

# Mostrar Serie con el nombre poblacion
poblacion_s
````

````python
# Mirar atributos de Series
print('Colección de valores:',poblacion_s.values)
print('Colección de etiquetas (indice):',poblacion_s.index)
````

Para crear un DataFrame desde un diccionario, tener en cuenta que las claves son las etiquetas de las columnas y los valores son los valores de las colecciones asociadas. Para el ejemplo son los valores de las series creadas previamente.

````python
# Crear Dataframe
provincia_df = pd.DataFrame( {'Población':poblacion_s, 'Superficie(km2)':superficie_s}  )
````

````python
# Mostrar
provincia_df
````

Las filas también se pueden recuperar por posición o nombre mediante un par de métodos. Usando las posiciones y el método `iloc` o las etiquetas y el método `loc`.

`iloc` se basa principalmente en la posición de números enteros pero también se puede utilizar con un array booleano.

**Indexación de sola las filas**

````python
# Número entero - escalar
# Fila en la posicion 0
provincia_df.iloc[0]
````

````python
# Lista de números enteros [0]
provincia_df.iloc[[0]]
````

````python
# Lista de números enteros [0,2] , filas de la posición 0 y posición 3

provincia_df.iloc[[0,3]]
````

````python
# Con un objeto de corte (slice)
#  [ini:fin:pasos]
provincia_df.iloc[:3:]
````

````python
# Con una máscara booleana de la misma longitud que el índice. En este caso indice es para las 5 filas
provincia_df.iloc[[True, False, False, False, False]]
````

**Indexación por fila y columna**

Puede combinar los tipos de indexadores para el índice y las columnas.

````python
# Datos de la fila en posición 0 y columna en posición 1
provincia_df.iloc[0 , 1]
````

````python
# Datos de las filas en posición 0 y 2, columna en posición 1
provincia_df.iloc[[0,2] , 1]
````

````python
# Con objeto de corte filas y columnas [ini:fin:pasos , ini:fin:pasos]
provincia_df.iloc[:3, :1]
````

`loc` permite el acceso a un grupo de filas y columnas por etiqueta(s) o un array booleano.

````python
# Ver df
provincia_df
````

````python
# Etiqueta de fila (indice)
provincia_df.loc['Salta']
````

````python
# Lista de etiquetas de filas
provincia_df.loc[['Salta','Buenos Aires']]
````

````python
# Etiqueta de fila y columna [ etiqueta fila (indice) , etiqueta columna ]
provincia_df.loc['Salta' ,'Población']
````

````python
# Por lista de etiquetas
provincia_df.loc[['Salta','Buenos Aires'], ['Población']]
````

````python
# Por objeto de corte
provincia_df.loc[:'Salta':, :'Población']
````

````python
# Lista booleana con la misma longitud que el eje de filas o columnas
provincia_df.loc[[True, False,False,False,False],[True,False]]
````

* La principal diferencia que existe entre loc e iloc es que en loc se usan las etiquetas (los nombres asignados tanto a las filas como a las columnas) mientras que en iloc se usan las posiciones de los elementos (la posición en la fila o la columna, comenzado a contar en 0).

Podemos indexar un DataFrame usando loc y máscaras booleanas

````python
# Filtro por operador relacional. 'Superficie(km2)' > 148008
provincia_df['Superficie(km2)'] > 148008
````

````python
# Por máscara
provincia_df.loc[provincia_df['Superficie(km2)'] > 148008, ['Población', 'Superficie(km2)']]
````

````python
provincia_df.loc[provincia_df['Superficie(km2)'] > 148008, ['Población']]
````

Otros métodos para la selección de valores que devuelve un subconjunto de los datos son `where`, `mask`, y `query`.

````python
# where() para reemplazar valores donde la condición sea Falso.
# provincia_df['Superficie(km2)'] > 148008 si es falso los valores serán 0
provincia_filtro_df1 = provincia_df.where(provincia_df['Superficie(km2)'] > 148008, 0 )

# Mostrar
provincia_filtro_df1
````

````python
# mask reemplaza los valores donde la condición sea Verdadera.
# provincia_df['Superficie(km2)'] > 148008 si es falso los valores serán 0
provincia_filtro_df2 = provincia_df.mask(provincia_df['Superficie(km2)'] > 148008, 0 )

# Mostrar
provincia_filtro_df2
````

````python
# query para consultar las columnas de un DataFrame con una expresión booleana.
provincia_df.query('`Superficie(km2)` > 148008') # es igual a provincia_df[provincia_df['Superficie(km2)']> 148008]
````

`map` aplica una función a un DataFrame elemento por elemento.
Este método aplica una función que acepta y devuelve un escalar a cada elemento de un DataFrame.

Podemos hacerlo utilizando funciones "lambda". Se utilizan para crear funciones anónimas. La expresión

```
lambda argumento(s) : expresion
```
Genera un objeto de función. El objeto sin nombre se comporta como un objeto de función definido con:

```
def <lambda>(parametros):
    return expresion
```
Para más información sobre `lambda`, click en [expresiones lambda en documentación oficial](https://docs.python.org/3/reference/expressions.html#lambda)


````python
# con funciones lambda.
provincia_df.map(lambda x: round(x,2))
````

````python
# También es posible utilizar el mapa con funciones que no son funciones lambda:
provincia_df.map(round, ndigits=1)
````

Los objetos que se pasan a la función son objetos Series cuyo índice es el índice del DataFrame (axis=0) o las columnas del DataFrame (axis=1).

Para más información, click en [apply en documento oficial](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)

````python
# Numpy
import numpy as np

# Aplicar suma por eje fila, axis = 0.
provincia_df.apply(np.sum, axis=0)
````

````python
# Aplicar suma por eje fila, axis = 1.
provincia_df.apply(np.sum, axis=1)
````

### ¿Cómo crear nuevas columnas derivadas de las columnas existentes en DataFrame?


Hasta ahora hemos guardados selecciones de filas o columnas en variables. Por ejemplo la selección de subconjunto de datos en `maxPun_vino_df` o `pais_vino_df`.

Recordemos, que cada columna en un DataFrame es una serie, por lo que podemos llamar a atributos y métodos para generar nuevos valores y almacenarlos como una nueva columna en el DataFrame.
Ahora veremos como crear una nueva columna en el DataFrame que llamaremos `price_stan`.



````python
# Crear nueva columna. Guardar valores de precio modificado
rev_vino_df['stand_price'] = (rev_vino_df['price'] - rev_vino_df['price'].mean()) /  rev_vino_df['price'].std()

# Mostrar las 5 primeras filas con las columnas price y stand_price'
rev_vino_df[['price','stand_price']].head()
````

Se desea cambiar el nombre de las columnas de datos del DataFrame se puede usar `rename`.

````python
# rename tiene el atributo inplace, por defecto es Falso, que da lugar a crear uno nuevo.
# Si es Verdadero, se ignora el valor de la copia, por lo tanto se modifica el Dataframe original
rev_vino_df.rename(
    columns={
        "stand_price": "Standard_price"
    }
)
````

### Tipos de Variables

En estadística las variables, las que se pueden variar, se clasifican en diferentes tipos según la naturaleza de los datos que representan. Aquí un resumen con ejemplos:

**Variables Cualitativas (o Categóricas)**

Son aquellas que representan características o categorías que no pueden ser medidas con números. Se dividen en dos tipos:

* Nominales: No tienen un orden específico.
  * Ejemplo: Color de ojos (azul, verde, marrón), estado civil (soltero, casado, divorciado).
* Ordinales: Tienen un orden o jerarquía, pero las diferencias entre valores no son necesariamente iguales.
  * Ejemplo: Nivel de satisfacción (bajo, medio, alto), nivel educativo (primaria, secundaria, universidad).

**Variables Cuantitativas (o Numéricas)**

Son aquellas que representan cantidades, medidas con número y permiten realizar operaciones matemáticas. Se dividen en dos tipos:

* Discretas: Toman un valor finito de valores. Son valores enteros contables (no pueden tomar valores intermedios).
  * Ejemplo: Número de integrantes en el hogar (0, 1, 2, 3), cantidad de autos en un estacionamiento.
* Continuas: Pueden tomar cualquier valor dentro de un rango númerico, incluyendo decimales.
  * Ejemplo: Altura de una persona (1.75 m), temperatura en grados Celsius (23.5°C).


En visualización, el tipo de variable determina la gráfica adecuada para su representación.



###  Tipos de gráficas  

Existen diversas formas de representar datos de manera visual. Algunas de las más utilizadas son:

* Gráficos de barras: Ideales para comparar valores de variables categóricas.

* Histogramas: Útiles para mostrar la distribución de una variable numérica, es decir cómo se distribuyen los diferentes valores de esa variable a lo largo de las observaciones.

* Diagramas de dispersión (scatter plots): Se emplean para analizar la relación entre dos variables numéricas.

* Gráficos de líneas: Representan tendencias a lo largo del tiempo.

* Boxplots: Muestran la distribución y posibles valores atípicos en un conjunto de datos.




### ¿Cómo crear gráficos con pandas?
Pandas permite representar gráficamente los datos de forma inmediata, utilizando el poder de la libreria Matplotlib, que estaremos explorando en el modulo de visualización.
Una manera sencilla de generar gráficos a partir de DataFrames, es utilizando el método `.plot`. Los métodos .plot.* son aplicables tanto a `Series` como a `DataFrame`.

A continuación, un ejemplo práctico, para ello consideremos el conjunto de datos de reseña de vino, visto en anteriores secciones, pero que corresponden a Argentina:

````python
# Seleccionar vinos de Argentina
rev_vino_argen_df = rev_vino_df[rev_vino_df['country']=='Argentina']

# Mostrar
rev_vino_argen_df.head(4)
````

````python
# Recordar tipo de datos del conjunto de datos del DataFrame
rev_vino_argen_df.dtypes
````

Una comprobación visual rápida de los datos con `.plot()`


````python
rev_vino_argen_df.plot()
````

Con un DataFrame, pandas crea de forma predeterminada un gráfico lineal para cada una de las columnas con datos numéricos.

Además del gráfico de líneas predeterminado al utilizar la función `plot`, hay varias alternativas disponibles para representar gráficamente los datos.

Cambiando el parámetro `kind`, podemos generar otros tipos de gráficos como `line` (líneas), `hist` (histograma), `box`(boxplot) o `scatter` (dispersión). Veamos ejemplos:

````python
# Explorar la distribución de variable númerica price mediante histograma
rev_vino_argen_df['price'].plot(kind='hist')
````

````python
# Explorar la relación de las variables númericas price y points. Usamos la gráfica scatter
# Solo se puede aplicar en DataFrame
rev_vino_argen_df[['price','points']].plot(kind='scatter', x='price', y='points')
````

````python
# Calculo la frecuencia de cada categoria de la variable categórica `Province`
# value_counts() devuelve una serie
provincias_s = rev_vino_argen_df['province'].value_counts()

# Mostrar
provincias_s
````

````python
# Gráfico de barra
provincias_s.plot(kind="bar")
````
